```python
import yfinance as yf
import pandas as pd

tesla = yf.download('TSLA')
print(tesla)
```

```python
tesla['High'].plot(figsize=(12,4))
```

```python
## xlimit and y limit
tesla['High'].plot(xlim=['2020-01-01','2021-09-01'],figsize=(12,4))
```

```python
tesla['High'].plot(xlim=['2020-01-01','2021-09-01'],ylim=[0,900],figsize=(12,4))
```

```python
tesla['High'].plot(xlim=['2020-01-01','2021-09-01'],ylim=[0,900],figsize=(12,4),ls='--',c='green')
```

```python
index= tesla.loc['2020-01-01':'2021-09-01'].index
share_open = tesla.loc['2020-01-01':'2021-09-01']['Open']
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
figure,axis=plt.subplots()
plt.tight_layout()
## Preventing overlapping
figure.autofmt_xdate()
axis.plot(index,share_open)
```

Time Resampling


```python
tesla.head()
```

```python
tesla.resample(rule='A').min()
```

```python
tesla.resample(rule='A').max()
```

```python
##year end frequency
tesla.resample(rule='A').max()['Open'].plot()
```

```python
tesla.resample(rule='QS').max()['High'].plot()
```

```python
tesla.resample(rule='BA').max()
```

```python
tesla.resample(rule='BQS').max()
```

```python
tesla['Open'].resample(rule='A').mean().plot(kind='bar')
```

```python
tesla['Open'].resample(rule='M').max().plot(kind='bar',figsize=(15,6))
```

```python
tesla['High'].rolling(11).max().head(20)
```

```python
tesla['Open:30 days rolling']=tesla['Open'].rolling(30).mean()
```
